```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

## Data Preprocessing

```python
tk=Tokenizer()
```

```python
tk.fit_on_texts([data])
```

```python

len(tk.word_index)
```

```python
input=[]
for se in data.split('\n'):
  sentence=tk.texts_to_sequences([se])[0]

  for i in range(1,len(sentence)):
    input.append(sentence[:i+1])
```

```python
input
```

```python
maxlen=max(len(x) for x in input )
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
pad=pad_sequences(input,maxlen=maxlen,padding='pre')
```

```python
pad
```

```python
x=pad[:,:-1]
```

```python
y=pad[:,-1]
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=549)
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,Embedding
```

## Model Building

```python
model=Sequential()
model.add(Embedding(549,100,input_length=77))
model.add(SimpleRNN(150))
model.add(Dense(549,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=100)
```

## Prediction
